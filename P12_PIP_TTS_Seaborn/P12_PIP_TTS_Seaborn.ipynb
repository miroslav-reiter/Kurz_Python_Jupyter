```python
import pyttsx3

# sapi5 - windows
# nsss - mac os/linux
# espeak - lubovolny os/plaformu

engine = pyttsx3.init(driverName="sapi4")
rate = engine.getProperty("rate")
voices = engine.getProperty("voices")

print("Rychlost: ", rate)

engine.setProperty("rate", 150)
# 0 Muz, 1 Zena
engine.setProperty("voice", voices[0].id)
engine.say("Hello, i am future Python developer")
engine.runAndWait()
```

```python
from gtts import gTTS

tts = gTTS("Hello, i am future Python developer")
tts.save("developer.mp3")
```

```python
import seaborn as sns

kvety = sns.load_dataset("iris")

sns.lineplot(x = "sepal_length", y="sepal_width", data=kvety)
sns.set_style("dark")
```

```python
import seaborn as sns

sns.set_theme()
tringelty = sns.load_dataset("tips")

sns.relplot(
    data = tringelty,
    x = "total_bill",
    y = "tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size"
)
```
