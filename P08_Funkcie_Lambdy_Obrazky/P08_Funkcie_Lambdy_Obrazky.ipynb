# 01. Funkcie

```python
import math
# Vytvorenie funkcie
# Funkcia = Definicia = Metoda (Funkcia v triede/class) = Operacie
def vypis_ahoj():
    """Funkcia, ktora ta pozdravi..."""
    print("Ahoj...")
    
# Volanie funkcie
print(type(vypis_ahoj))
vypis_ahoj()
print(vypis_ahoj.__doc__)

def vypocitaj_objem_valec(r = 1, v = 1):
    """Funkcia na vypocej objemu valca
       r - polomer v m default 1
       v - vyska v m default 1
    """
#     print("Objem valca je:", math.pi * math.pow(r,2) * v)
    objem = math.pi * math.pow(abs(r), 2) * abs(v)
#     print("Objem valca je:", round(objem, 2), "m^3")
    return round(objem, 2)

def vypocitaj_objem_gula(r = 1):
    """Funkcia na vypocej objemu gula
       r - polomer v m default 1
       Osetrenie na zaporne hodnoty
    """
    if isinstance(r, str):
        r = float(r)
    elif isinstance(r, int):
        r = float(r)
    
    return ((4/3) * math.pi * math.pow(r, 3))

        
print(vypocitaj_objem_valec.__doc__)
print(vypocitaj_objem_valec(10, 10))
print(vypocitaj_objem_valec())
print(vypocitaj_objem_valec(10))
print(vypocitaj_objem_valec(-10, -10))
print(vypocitaj_objem_valec(5, 2))

print(vypocitaj_objem_gula(10))
print(vypocitaj_objem_gula("   10   "))

# Mutovatelnost
# Immutable (Nemenitelne/nemutovatelne)
# 1. xreiterm@organizacia.domena
# 2. miroslav.reiter@organizacia.domena
def vytvor_firemny_email(m, p, domena = "abc.sk"):
    email2 = m.strip() + "." + p.strip() + "@" + domena
    email2 = email2.lower()
    
    m = m.strip()
    email1 = "x" + p.strip() + m[0] +  "@" + domena
    email1 = email1.lower()
    return email1

print(vytvor_firemny_email("  miroslav  ","  reiter  "))
meno = "  Adam  "
priezvisko = "  Sangala   "
print(vytvor_firemny_email(meno, priezvisko))
```

# 02. Trim/strip - Zbavenie sa medzier

```python
vstup = "   Karol   Sangala   "
print(vstup)
print(type(vstup))
# Funkcia len (length) - dlzka retazca - pocet znakov
print(len(vstup))
# strip zbav sa medzier na zaciatku aj na konci
print(vstup.strip())
print(len(vstup.strip()))

# rstrip zbav sa medzier na konci/napravo
# Trailing whitespace
print(vstup.rstrip())
print(len(vstup.rstrip()))

# lstrip zbav sa medzier na zaciatku/nalavo
# Leading whitespace
print(vstup.lstrip())
print(len(vstup.lstrip()))

vstup2 = " --  Karol ..  Sangala ,, **  "
print(vstup2.strip("-.,* "))
```

# 03. Overovanie retazcov, isinstance

```python
# isinstance
print("Je Karol retazec/string -->", isinstance("Karol", str))
print("Je 10 cele cislo/int -->", isinstance(10, int))
print("Je 10 retazec/string -->", isinstance(10, str))
print("Je 10.5 realne cislo/float -->", isinstance(10.5, float))
print("Je 10 realne cislo/float -->", isinstance(10, float))

# type
print("\nJe Karol retazec/string -->", type("Karol") == str)
print("Je 10 cele cislo/int -->", type(10) == int)
print("Je 10.5 realne cislo/float -->", type(10.5) == float)

print("\nJe Karol cislo/int -->", type("Karol") == int)
print("Je 10 cele retazec/string -->", type(10) == str)
print("Je 10.5 realne cislo/int -->", type(10.5) == int)
```

# 05. Lambda

Kratka/mala anonymna funkcia

```python
x = lambda a: a**2
print(x(5))

x = lambda a, b, c: a+b+c
print(x(1,2,3))
print(x(4,5,10))

def mocnina(a):
    return lambda q: q**a

druha_mocnina = mocnina(2)
tretia_mocnina = mocnina(3)
print(druha_mocnina(10))
print(tretia_mocnina(10))
```
